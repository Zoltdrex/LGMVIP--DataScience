```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

```python
img = cv2.imread("Beagle.jpg")
```

```python
cv2_imshow(img)
```

```python
#IMAGE COVERSION FROM COLOR TO B/W
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_grey)
```

```python
#IMAGE NEGATIVE
img_neg = cv2.bitwise_not(img_grey)
cv2_imshow(img_neg)
```

```python
#IMAGE SMOOTHENING
img_smooth = cv2.GaussianBlur(img_neg, (21,21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smooth)
```

```python
def dodgeV2(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
#FINAL SKETCH IMAGE
final_img = dodgeV2(img_grey, img_smooth)
cv2_imshow(final_img)
```

```python
#FINAL RESULT
cv2_imshow(img)
cv2_imshow(final_img)
```
